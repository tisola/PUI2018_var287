```python
from __future__ import print_function
__author__= 'var287'

import numpy as np
import pylab as pl

%pylab inline
```

```python
np.random.seed(2018)
```

```python
## Set Mean
lam= 150 
```

#### Generating samples

```python
norm = []
pois = []
binom =[]
chisq =[]
logis =[]

for i in range(100):
    s = int(np.random.rand()*1990+10) ## create size
    norm.append(np.random.normal(lam,1, s))
    pois.append(np.random.poisson(lam, s))
    binom.append(np.random.binomial((lam*2),0.5, s))
    chisq.append(np.random.chisquare(lam, s))
    logis.append(np.random.logistic(lam, 1, s))
```

#### Plotting sample mean against sample size

```python
fig = plt.figure(figsize=(10,10))
fig.subplots_adjust(hspace=0.6, wspace=0.4)

## gaussian
nx = fig.add_subplot(3,2,1)

for i in norm:
    nx.plot(len(i),i.mean(), '.')

nx.plot([0,2000], [150,150], '-')
nx.set_title("Gaussian")
nx.set_xlabel('Sample Size', fontsize = 10)
nx.set_ylabel('Sample Mean', fontsize = 10)


## Poisson 
px = fig.add_subplot(3,2,2)
for i in pois:
    px.plot(len(i),i.mean(), '.')
    
px.plot([0,2000], [150,150], '-')
px.set_xlabel('Sample Size', fontsize = 10)
px.set_ylabel('Sample Mean', fontsize = 10)
px.set_title("Poisson")


## Binomial
bx = fig.add_subplot(3,2,3)

for i in binom:
    bx.plot(len(i),i.mean(), '.')
    
bx.plot([0,2000], [150,150], '-')
bx.set_title("Binomial")
bx.set_xlabel('Sample Size', fontsize = 10)
bx.set_ylabel('Sample Mean', fontsize = 10)


## Chi-Square
cx = fig.add_subplot(3,2,4)

for i in chisq:
    cx.plot(len(i),i.mean(), '.')
    
cx.plot([0,2000], [150,150], '-')
cx.set_title("Chi-Square")
cx.set_xlabel('Sample Size', fontsize = 10)
cx.set_ylabel('Sample Mean', fontsize = 10)


## Logistic
lx = fig.add_subplot(3,2,5)

for i in logis:
    lx.plot(len(i),i.mean(), '.')

lx.plot([0,2000], [150,150], '-')
lx.set_title("Logistic")
lx.set_xlabel('Sample Size', fontsize = 10)
lx.set_ylabel('Sample Mean', fontsize = 10)

```

Figure 1: Plots of sample means against sample size generated from Gaussian, Poisson, Binomial, Chi-Square, and Logistic random sampling with population mean of 150. The general trend of all of the plots proves the Law of Large Number, where the mean of the sample gets closer to the population mean of 150 as the sample size increases.

#### Distribution of Sample Means

```python
fig2 = plt.figure(figsize=(10,10))
fig2.subplots_adjust(hspace=0.6, wspace=0.4)

## Normal
nh = fig2.add_subplot(3,2,1)
n = []
for i in range(100):
    n.append(norm[i].mean())

nh.hist(n,bins=30)
nh.set_title('Gaussian')
nh.set_xlabel('sample mean', fontsize = 10)
nh.set_ylabel('N', fontsize = 10)


## Poisson
nh = fig2.add_subplot(3,2,2)
n = []
for i in range(100):
    n.append(pois[i].mean())

nh.hist(n,bins=30)
nh.set_title('Poisson')
nh.set_xlabel('sample mean', fontsize = 10)
nh.set_ylabel('N', fontsize = 10)


## Binomial
nh = fig2.add_subplot(3,2,3)
n = []
for i in range(100):
    n.append(binom[i].mean())

nh.hist(n,bins=30)
nh.set_title('Binomial')
nh.set_xlabel('sample mean', fontsize = 10)
nh.set_ylabel('N', fontsize = 10)


## Chi-Square
nh = fig2.add_subplot(3,2,4)
n = []
for i in range(100):
    n.append(chisq[i].mean())

nh.hist(n,bins=30)
nh.set_title('Chi-Square')
nh.set_xlabel('sample mean', fontsize = 10)
nh.set_ylabel('N', fontsize = 10)


# Logistic
nh = fig2.add_subplot(3,2,5)
n = []
for i in range(100):
    n.append(logis[i].mean())

nh.hist(n,bins=30)
nh.set_title('Logistic')
nh.set_xlabel('sample mean', fontsize = 10)
nh.set_ylabel('N', fontsize = 10)
```

Figure 2: Histograms of sample means distribution generated from Gaussian, Poisson, Binomial, Chi-Square, and Logistic random sampling with population mean of 150. The histograms supports the Central Limit Theorem as the all of the sample means distributions in the histograms form a gaussian distribution that centers around the mean of 150.
